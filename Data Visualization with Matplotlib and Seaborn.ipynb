## Plotting Pairwise Relationships

You'll find it helpful to plot pairwise relationships between multiple numeric variables. For e.g., here we have taken the prices of some popular cryptocurrencies such as bitcoin, litecoin, ethereum, monero, neo, quantum and ripple. 

Now, the crypto enthusiasts would know that the prices of these currencies vary with each other. If bitcoin goes up, the others will likely follow suit, etc.


Now, say you want to trade in some currencies. Given a set of cryptocurrencies, how will you decide when and which one to buy/sell? It will be helpful to analyse past data and identify some trends in these currencies.




````python
# reading cryptocurrency files
btc = pd.read_csv("C:\\Users\\gauri\\OneDrive\\Documents\\Extra\\Machine Learning Kaggle course\\7_DataVisualization\\Datavisual\\Datavisual\\crypto_data\\bitcoin_price.csv")
ether = pd.read_csv("C:\\Users\\gauri\\OneDrive\\Documents\\Extra\\Machine Learning Kaggle course\\7_DataVisualization\\Datavisual\\Datavisual\\crypto_data\\ethereum_price.csv")
ltc = pd.read_csv("C:\\Users\\gauri\\OneDrive\\Documents\\Extra\\Machine Learning Kaggle course\\7_DataVisualization\\Datavisual\\Datavisual\\crypto_data\\litecoin_price.csv")
monero = pd.read_csv("C:\\Users\\gauri\\OneDrive\\Documents\\Extra\\Machine Learning Kaggle course\\7_DataVisualization\\Datavisual\\Datavisual\\crypto_data\\monero_price.csv")
neo = pd.read_csv("C:\\Users\\gauri\\OneDrive\\Documents\\Extra\\Machine Learning Kaggle course\\7_DataVisualization\\Datavisual\\Datavisual\\crypto_data\\neo_price.csv")
quantum = pd.read_csv("C:\\Users\\gauri\\OneDrive\\Documents\\Extra\\Machine Learning Kaggle course\\7_DataVisualization\\Datavisual\\Datavisual\\crypto_data\\qtum_price.csv")
ripple = pd.read_csv("C:\\Users\\gauri\\OneDrive\\Documents\\Extra\\Machine Learning Kaggle course\\7_DataVisualization\\Datavisual\\Datavisual\\crypto_data\\ripple_price.csv")
````

Since all the datasets have similar column names, we add few initials to it in order to differentiate when it all comes together

````python
btc.columns = btc.columns.map(lambda x: str(x) + '_btc')
ether.columns = ether.columns.map(lambda x: str(x) + '_et')
ltc.columns = ltc.columns.map(lambda x: str(x) + '_ltc')
monero.columns = monero.columns.map(lambda x: str(x) + '_mon')
neo.columns = neo.columns.map(lambda x: str(x) + '_neo')
quantum.columns = quantum.columns.map(lambda x: str(x) + '_qt')
ripple.columns = ripple.columns.map(lambda x: str(x) + '_rip')

btc.head()
````

We merge all the tables using common column of Date

````python
# merging all the files by date
m1 = pd.merge(btc, ether, how="inner", left_on="Date_btc", right_on="Date_et")
m2 = pd.merge(m1, ltc, how="inner", left_on="Date_btc", right_on="Date_ltc")
m3 = pd.merge(m2, monero, how="inner", left_on="Date_btc", right_on="Date_mon")
m4 = pd.merge(m3, neo, how="inner", left_on="Date_btc", right_on="Date_neo")
m5 = pd.merge(m4, quantum, how="inner", left_on="Date_btc", right_on="Date_qt")
crypto = pd.merge(m5, ripple, how="inner", left_on="Date_btc", right_on="Date_rip")

crypto.head()
````

````python
# Subsetting only the closing prices column for plotting
curr = crypto[["Close_btc", "Close_et", 'Close_ltc', "Close_mon", "Close_neo", "Close_qt"]]
curr.head()
````

````python
sns.pairplot(curr)
plt.show()
````

````python
a = curr.corr()
a.round(3)
#if they are hgihly correlated then the values will be increasing
````

### Heatmaps

````python
# figure size
plt.figure(figsize=(10,8))

# heatmap
sns.heatmap(a, cmap="YlGnBu", annot=True)
plt.show()
````

The dark blue boxes show the most correlated currencies. Specifically, **ethereum-quantum (0.79)** and **monero-ltc (0.79)** are the most correlated pairs. Also, **neo-btc (0.73)** are quite highly correlated.
